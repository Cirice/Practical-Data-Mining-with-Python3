```python
import networkx as nx
%matplotlib inline
```

```python
n = 50 # n nodes
```

```python
p = 0.1
```

```python
w = [p*n for i in range(n)]      # w = p*n for all nodes
```

```python
G = nx.expected_degree_graph(w)  # configuration model
```

```python
dh = nx.degree_histogram(G)
```

```python
low = min(d for n, d in G.degree())
```

```python
nx.draw(G)
```

```python
print"degree (#nodes) ****"
for i in range(low, len(dh)):
    bar=''.join(dh[i]*['*'])
    print i, dh[i], bar
```
